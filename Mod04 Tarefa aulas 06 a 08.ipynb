####  2) Melhorando a interação com o API
Observe dois URLs de consultas diferentes, por exemplo o URL utilizado em aula, e o URL feito no exercício anterior. Compare-os e observe as diferenças.

1. Faça uma função em Python que recebe como argumento o UF da consulta e o tipo de interessado (```'Estado'```ou ```Município```), e que devolve os dados da consulta no formato *DataFrame*.
2. Quantas solicitações para o Estado podem ser consultadas para Minas Gerais com *status* em 'Arquivado por decurso de prazo' estão registradas?
3. Qual é o município da Bahia com mais solicitações deferidas?

````python
url2 = 'https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl'
````

````python
r2 = requests.get(url2)
````

````python
r2.status_code
````

````python
data_json2 = r2.json()
````

````python
print(data_json2['items'])
````

````python
# 1) Faça uma função em Python que recebe como argumento o UF da consulta e o tipo de interessado ('Estado'ou Município)
def uf_tipo_interessado():
    print('Informar a UF desejada: ')
    df = pd.DataFrame(data_json2['items'])
    lista_uf = df['uf'].unique()
    print(lista_uf)
    uf = input().upper()
    print('A UF selecionada foi: ',uf)
    print('Informar a tipo de interessado desejado (Estado ou Município): ')
    lista_tipo_interessado = df['tipo_interessado'].unique()
    print(lista_tipo_interessado)
    tipo_interessado = input().capitalize()
    print('O tipo de interessado selecionado foi: ',tipo_interessado)
    df = pd.DataFrame(data_json2['items'])
    if uf in lista_uf:
        if tipo_interessado in lista_tipo_interessado:
            cond = (df['uf'] == uf) & (df['tipo_interessado'] == tipo_interessado)
            resultado = df[cond]
            print(uf,tipo_interessado)
            return resultado
        else:
            print('Tipo de interessado não presente na lista!')
            print('Refaça a consulta!')
    else:
        print('UF não presente na lista!')
        print('Refaça a consulta!')
````

````python
# obtendo os dados de UF e Tipo de interessado
df_resultado = uf_tipo_interessado()
````

````python
df_resultado.head()
````

````python
# 2) Quantas solicitações para o Estado podem ser consultadas para Minas Gerais com status em 'Arquivado por decurso de prazo'?
solicitacoes = df_resultado['status'] == 'Arquivado por decurso de prazo'
consulta = df_resultado[solicitacoes]['status'].sum()
print('Quantidade de solicitações: ',consulta)
````

````python
# 3) Qual é o município da Bahia com mais solicitações deferidas?
df_resultado = uf_tipo_interessado()
````

````python
df_resultado.head()
````

````python
# consultando os municípios e quantidade de solicitações deferidas
solicitacoes = df_resultado['status'] == 'Deferido'
consulta = df_resultado[solicitacoes]['interessado'].value_counts()
consulta.head(5)
````

````python
# 4) Salve um arquivo .csv com os dados de solicitações da Bahia, com interessado = 'Estado'
df_resultado = uf_tipo_interessado()
````
