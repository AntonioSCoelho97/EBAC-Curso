```python
# Importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Conectando o arquivo 'sinasc_RO_2019.csv'
sinasc = pd.read_csv('SINASC_RO_2019.csv')
```

```python
# Visualizando DataFrame 'sinasc'
sinasc.head()
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
# Visualizando os 5 gráficos
plt.rc('figure', figsize=(10, 8))
fig, axes = plt.subplots(3, 2)

sns.countplot(ax = axes[0, 0], x='APGAR1', data=sinasc)
sns.countplot(ax = axes[0, 1], x='APGAR5', data=sinasc)
sns.countplot(ax = axes[1, 0], x='SEXO', data=sinasc)
# Para uma melhor visualização foi invertido a apresentação do gráfico da variável 'GESTACAO'
sns.countplot(ax = axes[1, 1], y='GESTACAO', data=sinasc) 
sns.countplot(ax = axes[2, 0], x='ESTCIVMAE', data=sinasc)

#ajustar o espaçamento
plt.subplots_adjust(wspace=0.7, hspace=0.3)
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
# Apresentando o APGAR5 médio por CONSULTAS
plt.close('all')
sns.barplot(x='CONSULTAS', y='APGAR5', data=sinasc)
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
# Visualizando as variáveis SEXO, GESTACAO, ESTCIVMAE e CONSULTAS
plt.close('all')
plt.rc('figure', figsize=(10, 8))
fig, axes = plt.subplots(2, 2)

sns.countplot(ax = axes[0, 0], x='SEXO', data=sinasc)
# Para uma melhor visualização foi invertido a apresentação do gráfico da variável 'GESTACAO'
sns.countplot(ax = axes[0, 1], y='GESTACAO', data=sinasc)
sns.countplot(ax = axes[1, 0], x='ESTCIVMAE', data=sinasc)
sns.countplot(ax = axes[1, 1], x='CONSULTAS', data=sinasc)

#ajustar o espaçamento
plt.subplots_adjust(wspace=0.7, hspace=0.3)
```
