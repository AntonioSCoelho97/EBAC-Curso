#### 1. Montando um passo a passo para o Bagging

1. escolha o algoritmo de machine learning para aplicação do Bagging: árvore de decisão, árvore de regressão, regressão logística, etc;
2. ajuste a base de dados ao algoritmo escolhido;
3. separe as bases de treino para o Bagging: através de sorteio com reposição;
3. aplique o algoritmo escolhido em cada base de treino, gerando vários modelos; e
4. realize a votação caso o algoritmo seja de classificação ou a média caso seja um algoritmo de regressão.

#### 2. Explicando com suas palavras o Bagging

A Bagging é uma das técnicas de ensemble learning que pode ser combinado a vários algoritmos de machine learning para torná-los mais genéricos, evitando que o modelo gerado fique viciado à base de treino; melhorando a performance do algoritmo pois dificulta a ocorrência de overfitting. 
O algoritmo escolhido é treinado nas diversas bases de treino, geradas por amostragem aleatória com reposição, criando diversos modelos que tornarão o resultado final mais genérico. 
A agregação ocorre por votação para os algoritmos de classificação ou pela média para os de regressão.

#### 3. Implementando em python o código do Bagging 

```python
# Importando as bibliotecas
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.tree import plot_tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier

from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay

from sklearn.model_selection import train_test_split
```

```python
# Carregando base de dados 'demo01.csv'
df = pd.read_csv('demo01_dummy.csv')
df_linhas, df_col = df.shape   # Consultando o número de linhas e colunas
print('A base de dados tem', df_linhas,'linhas e', df_col,'colunas')
df.head()
```

```python
# Excluíndo coluna "Unnamed: 0"
df.drop("Unnamed: 0",axis = 1, inplace = True)
df.head()
```

```python
# Ajustando a coluna 'mau' para do tipo 'inteiro'
df.mau = df.mau.astype(int)
df.head()
```

```python
# Separando as variáveis
y = df.mau
X = df.drop('mau',axis=1)
```

```python
# Separando 70% da base para treinamento e 30% para validação
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100)
```

```python
# Definindo um objeto com a função da árvore de decisão
clf = DecisionTreeClassifier(random_state=100)
```

```python
# Criando um Bagging
bagging_clf = BaggingClassifier(clf, n_estimators=100, random_state=100)
```

```python
# Treinando os modelos com os dados separados para treinamento
bagging_clf.fit(X_train, y_train)
```

#### A partir daqui vamos apresentar os resultados individuais para melhor compreensão do Bagging

```python
# Obtendo os estimadores individuais.
modelo_clf = bagging_clf.estimators_
```

```python
# Separando uma linha da base de teste
X_teste = X_test.iloc[0].to_frame().transpose()
```

```python
# Ajustando a linha para aplicação do predict
X_teste_modelo = X_teste.values
```

```python
# Gerando a lista das previsões de cada modelo
lista = []
for i, modelo in enumerate(modelo_clf):
    y_pred = modelo.predict(X_teste_modelo)
    lista.append(y_pred)
```

```python
# Gerando DataFrame com as previsões
y_lista = pd.DataFrame(lista).rename(columns={0: 'previsão'})
y_lista.head()
```

```python
# Identificando a maior ocorrência
y_lista.value_counts()
```

```python
# Verificando o resultado da previsão sem a separação individual, 
# tendo em vista que a implementação 'BaggingClassifier' já faz todo o trabalho.
y_pred_bagging = bagging_clf.predict(X_teste)
y_pred_bagging
```

**Análise:** o resultado obtido foi '0', a maior ocorrência observada nos modelos.

```python
# Verificando a acurácia do bagging para a base de treino
predict_train = bagging_clf.predict(X_train)
accuracy_score(y_train, predict_train)
```

```python
# Verificando a acurácia do bagging para a base de teste
predict = bagging_clf.predict(X_test)
accuracy_score(y_test, predict)
```

```python
# Visualizando a matriz de classificação
predictions = bagging_clf.predict(X_test)
cm = confusion_matrix(y_test, predictions, labels=bagging_clf.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Aprovados', 'Reprovados'])
disp.plot()
```
