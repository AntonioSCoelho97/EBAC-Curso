####  3) Faça um histograma da idade da mãe

```python
# Fazendo um histograma da idade da mãe
sns.displot(data = sinasc
            ,x = 'IDADEMAE'    # Coluna PESO no eixo x
            ,bins = 20         # Passando o número de compartimentos
            ,element = 'step'  # Retirando as linhas internas do histograma
            ,kde = True        # Visualizando a Densidade de Probabilidade de uma variável contínua.
            ,rug = True        # Visualizando um gráfico de "tapete"
           )
```
