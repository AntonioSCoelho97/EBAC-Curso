## Exercício 1: Vestibular

Considere que a os dados gerados na célula abaixo contêm o número de acertos de 100 alunos em um vestibular para um curso de exatas, divididas pelos respectivos assuntos. Considere que cada assunto possui um número de questões conforme a tabela abaixo:

| assunto | número de questões |
|:---:|:---:|
| Matemática | 24 |
| Português | 18 |
| Geografia | 8 |
| Inglês | 8 |
| História | 8 |
| Física | 12 |
| Química | 12 |

Usando os comandos de operações com DataFrames que você aprendeu na Aula 03, calcule:

1. (operações com escalar) Calcule o percentual de acerto dos alunos por assunto.  
2. (operações entre *DataFrames) Calcule o total de acertos de cada aluno.  
3. Calcule o porcentual geral de cada aluno.  
4. Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?  

````python
import pandas as pd
import numpy as np

np.random.seed(42)
df_mat = pd.DataFrame(np.random.randint(24, size=(100, 1)), columns=['Qt_acertos'])

df_por = pd.DataFrame(np.random.randint(18, size=(100, 1)), columns=['Qt_acertos'])

df_geo = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_ing = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_his = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_fis = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])

df_qui = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])
````

````python
# 1) Seu código aqui
pd.options.display.float_format = '{:.2f}'.format # formatar os valores para duas casas decimais
pc_acerto_assunto = pd.DataFrame() # criar DataFrame vazio
# inserir colunas com o percentual de acerto dos alunos por assunto
pc_acerto_assunto['pc_mat (%)'] = 100 * df_mat / 24 
pc_acerto_assunto['pc_por (%)'] = 100 * df_por / 18 
pc_acerto_assunto['pc_geo (%)'] = 100 * df_geo /  8 
pc_acerto_assunto['pc_ing (%)'] = 100 * df_ing /  8 
pc_acerto_assunto['pc_his (%)'] = 100 * df_his /  8 
pc_acerto_assunto['pc_fis (%)'] = 100 * df_fis / 12 
pc_acerto_assunto['pc_qui (%)'] = 100 * df_qui / 12 
pc_acerto_assunto
````

````python
# 2) Seu código aqui
total = pd.DataFrame(columns = ['Qt_acerto']) # cria um dataframe vazio
total = df_mat + df_por + df_geo + df_ing + df_his + df_fis + df_qui # realizar operação entre dataframes
total
````

````python
# 3) Seu código aqui
pc_geral = pd.DataFrame() # criar um dataframe vazio
pc_geral['Geral (%)'] = 100 * total / 90 # inserir coluna com o percentual geral de cada aluno
pc_geral
````

````python
# 4) Seu código aqui
total.loc[total['Qt_acertos'] > 45, 'aprovados'] = 1 # criar coluna 'aprovados' para os alunos com nota acima de 45
total_gt_45 = total['aprovados'].sum() # realizar a soma dos alunos com nota acima de 45
total_gt_45
````

## 2) Vestibular II

Ainda sobre o mesmo banco de dados:

1. Neste vestibular, quem 'zera' em matemática, física ou química está desqualificado. Monte um novo *DataFrame* com os alunos desqualificados por este critério.
2. Quantos são esses alunos?
3. Qual a média desses alunos em história e geografia?
4. Monte um *DataFrame* com os alunos que passaram para a segunda fase. Repare que estes alunos não podem ter sido desqualificados.

````python
# seu código aqui
# DataFrame com os alunos zerados em matemática, física ou química
df_eliminados = total[(df_mat['Qt_acertos'] == 0) | 
                      (df_fis['Qt_acertos'] == 0) | 
                      (df_qui['Qt_acertos'] == 0)]

df_eliminados
````

````python
# quanto são estes alunos?
qtde_eliminados = df_eliminados.copy()
# criar coluna 'eliminados' para os alunos zerados em matemática, física ou química
qtde_eliminados.loc[df_eliminados['Qt_acertos'] > 0, 'eliminados'] = 1 
qtde_eliminados['eliminados'].sum()
````

````python
# Qual a média desses alunos em história e geografia?
qtde_eliminados['Nota_His'] = df_his
qtde_eliminados['Nota_Geo'] = df_geo
media_his = qtde_eliminados['Nota_His'].mean()
media_geo = qtde_eliminados['Nota_Geo'].mean()
print('As médias em história e geografia são: ',media_his, ' e ', media_geo)
````

````python
# Monte um DataFrame com os alunos que passaram para a segunda fase. Repare que estes alunos não podem ter sido desqualificados.
# criar coluna com a indicação dos eliminados por zerar matemática, física ou química
total['eliminados'] = qtde_eliminados['eliminados']
# filtrar os alunos que tiveram nota acima de 45 e não foram zerados em matemática, física ou química
df_qualificados = total[(total['aprovados'] == 1) & (total['eliminados'] != 1)]
df_qualificados
````
