````python
# Importando as Bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
# Criando DataFrame 'prev_renda'
prev_renda = pd.read_csv('previsao_de_renda.csv')
prev_renda.head()
````

````python
# Gerando o gráfico para a variável 'posse de veiculo'
ax = sns.pointplot(x="posse_de_veiculo", y="renda",
                   data=prev_renda, dodge=True, errorbar=('ci', 90))
````

````python
# Gerando o gráfico para a variável 'posse de veiculo' ao longo do tempo
ax = sns.pointplot(x="data_ref", y="renda", hue = 'posse_de_veiculo',
                   data=prev_renda, dodge=True, errorbar=('ci', 95))
ax.set_xticklabels(ax.get_xticklabels(), rotation = 45)

plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.30),ncol = 3)
````

````python
# Gerando o gráfico para a variável 'posse de imóvel'
ax = sns.pointplot(x="posse_de_imovel", y="renda",
                   data=prev_renda, dodge=True, errorbar=('ci', 90))
````

````python
# Gerando o gráfico para a variável 'posse de imóvel' ao longo do tempo
ax = sns.pointplot(x="data_ref", y="renda", hue = 'posse_de_imovel',
                   data=prev_renda, dodge=True, errorbar=('ci', 95))
ax.set_xticklabels(ax.get_xticklabels(), rotation = 45)

plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.30),ncol = 3)
````

## Análise
* A variável **posse_de_veiculo** parece discriminar renda de forma estatisticamente significante, pois apresenta **renda superior** para os clientes que **possuem veículos** ao longo do tempo; situação não observada para a variável **posse_de_imovel**, pois em algumas datas de referência os clientes que **não possuem imóveis** apresentam em média **renda superior**.
