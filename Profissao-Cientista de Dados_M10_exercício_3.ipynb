````python
# Importando Bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
# Criando DataFrame 'prev_renda'
prev_renda = pd.read_csv('previsao_de_renda.csv')
prev_renda.head()
````

#### Variável 'posse_de_veiculo'

````python
# Avaliando a distribuição da variável 'posse_de_veiculo'
ax = sns.countplot(x='data_ref',  hue = 'posse_de_veiculo', data=prev_renda)

ax.set_xticklabels(ax.get_xticklabels(), rotation = 45)

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
````
