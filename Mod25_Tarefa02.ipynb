#### 1- Calcule de forma num√©rica a integral da fun√ß√£o  ùë•3 avaliada entre os pontos 1 e 2. Identifique o n√∫mero de ret√¢ngulos que voc√™ usou.

```python
# Importando as bibliotecas
import numpy as np
import matplotlib.pyplot as plt

from matplotlib.animation import FuncAnimation, PillowWriter 
from IPython.display import HTML
import warnings

%matplotlib inline
```

```python
warnings.filterwarnings("ignore")

# seno de x
f = lambda x: x**3

def calcula_area(a, b, func, frame_num):
    num_bins=frame_num+1 # Quantidade de ret√¢ngulos
    delta = (b-a)/num_bins # largura do ret√¢ngulo
    x_cols = np.linspace(a, b-delta, num_bins)+delta/2 # Divis√£o do eixo X
    y_cols = f(x_cols) # alturas dos ret√¢ngulos
    return sum(delta*y_cols)
    

def animate(n, a=1, b=2, func=f, verdadeiro_valor = 3.75):
    
    plt.clf()  # Limpa o gr√°fico anterior para montar o pr√≥xim frame
    ax = plt.axes(xlim=(a - 0.1, b + 0.1), ylim = (0, 10))  # Monta o sistema de eixos
    
    # Plota a fun√ß√£o
    x = np.linspace(a ,b, 1000)
    y = f(x)
    ax.plot(x, y, '-');
    
    # Define as quebras e as alturas das colunas
    num_bins=n+1
    delta = (b-a)/num_bins
    x_cols = np.linspace(a, b-delta, num_bins)+delta/2
    y_cols = f(x_cols)

    # plota barras
    bar = ax.bar(x_cols, y_cols, alpha=.5, width = delta, edgecolor='blue')   

    # √Årea dos ret√¢ngulos
    soma_riemann = calcula_area(a, b, func, n)
    
    # Erro
    erro = verdadeiro_valor - soma_riemann
    ax.set_title(f'Soma de Riemann = {soma_riemann:.2f} erro = {erro:.4f}')

    return bar
animate(86);
```

**OBS**: a partir de 86 frames ou **87 ret√¢ngulos**, o valor do erro para 4 (quatro) casas de precis√£o √© **zerado**.

#### 2- Calcule essa √°rea de forma anal√≠tica.

$\int_{1}^{2} x^3 \, dx = \left[ \frac{{x^4}}{4} \right]_{1}^{2} = \frac{{2^4}}{4} - \frac{{1^4}}{4} = \frac{16}{4} - \frac{1}{4} = 4 - \frac{1}{4} = \frac{15}{4} = 3.75$
